This class is used to plot data in the form of [(xdata, ydata)] where xdata and ydata are lists of points to plot. This allows for multiple sets of data to be plotted simultaneously.When plotting, the top plot becomes interactive. The mouse can be used to choose a domain which is plotted in the second plot below the main one on top. This allows for 'zooming' in.

```python
import numpy as np
import matplotlib.pyplot as plt
import PlotClass as pc
# uncomment the following for pop-up interactive plot
# %matplotlib qt

xdata1 = np.arange(0, 10)
xdata2 = np.arange(10, 20)
xdata3 = np.arange(20, 30)
xdata4 = np.arange(30, 40)
ydata1 = np.arange(0, 10)
ydata2 = np.arange(10, 20)
ydata3 = np.arange(30, 40)
ydata4 = np.arange(5, 15)

data_list = [(xdata1, ydata1), (xdata2, ydata2), (xdata3, ydata3), (xdata4, ydata4)]

obj = pc.PlotClass(data_list, title="Example of PlotClass")
obj.onselect(9, 31) #remove to initialize both plots as same
plt.show()
```
